```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '../data/Part02/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
print(train_Y.shape)
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns; sns.set(color_codes=True)
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))

sns.regplot(x = df['1stFlrSF'], y=train_Y)
#sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
plt.figure(figsize=(8, 6))
df_1 = df

df_1['1stFlrSF'] = df_1['1stFlrSF'].clip(500, 2000)
sns.regplot(x = df_1['1stFlrSF'], y=train_Y)
print(df_1['1stFlrSF'].shape, df_1['1stFlrSF'].min(), df_1['1stFlrSF'].max())
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df_1)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
plt.figure(figsize=(8, 6))

keep_indexs = (df['1stFlrSF']> 500) & (df['1stFlrSF']< 2000)
df_2 = df[keep_indexs]
train_Y_2 = train_Y[keep_indexs]

sns.regplot(x = df_2['1stFlrSF'], y=train_Y_2)
print(df_2['1stFlrSF'].shape, df_2['1stFlrSF'].min(), df_2['1stFlrSF'].max())
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df_2)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y_2, cv=5).mean()
```

#### NOTE
作業一和作業二的差別在於，作業一只是把疑似outlier的值強制改成上下限，並沒有排除資料點，可以看到圖上有很多被強制更改為上下限的點在邊界，這樣沒有排除的情況會影響預測準確程度。作業二則是直接移除outlier來做預測，對資料分布預測就會比較準確。
